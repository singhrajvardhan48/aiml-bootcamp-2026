Calculate the avg marks and prin t class fialed or passed is the avg is 50 or above

```python
marks = [45, 60, 72, 38, 90]

total = 0

for i in range(len(marks) - 1):
    total += marks[i]

average = total / len(marks)

if average > 50:
    print("Class passed")
else:
    print("Class failed")

print("Average:", average)
```

```python
import pandas as pd
```

```python
data = {
    "student_id": [101, 102, 103, 104, 105, 106],
    "attendance_percent": [92, 67, 81, 45, 74, 88],
    "assignment_score": [18, 12, 15, 8, 14, 19],
    "quiz_score": [72, 48, 65, 30, 55, 80],
    "lab_completed": [True, False, True, False, True, True]
}

df = pd.DataFrame(data)

df
```

```python
df.shape
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df['total_score'] = df['assignment_score'] + df['quiz_score']
df
```

```python
df['eligible'] = (
    (df['attendance_percent'] >= 75) &
    (df['total_score'] >= 70) &
    (df['lab_completed'])
)
df
```

```python
eligible_students = df[df["eligible"] == True]
eligible_students
```

```python
import matplotlib.pyplot as plt

plt.hist(df['attendance_percent'])

plt.title("Attendance Distribution")
plt.xlabel("Attendance %")
plt.ylabel("Number of Students")

plt.show()
```

```python

df['total_score']
```

```python
plt.bar(
    df['student_id'],
    df['total_score']
)

plt.title("Student Total Scores")
plt.xlabel("Student ID")
plt.ylabel("Total Score")

plt.show()
```

```python
plt.scatter(
    df['attendance_percent'],
    df['total_score']
)

plt.title("Attendance vs Total Score")
plt.xlabel("Attendance %")
plt.ylabel("Total Score")

plt.show()
```

```python


df['performance_level'] = pd.cut(
    df['total_score'],
    bins=[0,50,75,100],
    labels=['Low','Medium','High']
)

df
```

```python
plt.pie(
    df['eligible'].value_counts(),
    labels=['Eligible', 'Not Eligible'],
    autopct='%1.1f%%'
)

plt.title('Eligibility Distribution')

plt.show()
```

```python

plt.scatter(
    df['attendance_percent'],
    df['quiz_score']
)

plt.title('Attendance vs Quiz Score')
plt.xlabel('Attendance %')
plt.ylabel('Quiz Score')

plt.show()
```

```python

plt.bar(
    df['student_id'],
    df['assignment_score'],
    label='Assignment'
)

plt.bar(
    df['student_id'],
    df['quiz_score'],
    bottom=df['assignment_score'],
    label='Quiz'
)

plt.title('Assignment + Quiz Contribution')
plt.xlabel('Student ID')
plt.ylabel('Score')

plt.legend()

plt.show()
```

```python


df['performance_level'].value_counts().plot(kind='bar')

plt.title('Performance Levels')
plt.xlabel('Level')
plt.ylabel('Number of Students')

plt.show()
```

```python
df['efficiency_score'] = (
    df['total_score'] / df['attendance_percent']
)
```

```python


plt.bar(
    df['student_id'],
    df['efficiency_score']
)

plt.title('Student Efficiency Score')
plt.xlabel('Student ID')
plt.ylabel('Efficiency')

plt.show()
```
